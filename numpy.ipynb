```python3
def in_prime(n):
  for i in range(2,n):
    if n%i==0:
      return False
    else:
      return True
out=[]
for i in range(2,101):
  if in_prime(i):
    out.append(i)
prime=pd.Series(out,name='Prime')
prime
```
